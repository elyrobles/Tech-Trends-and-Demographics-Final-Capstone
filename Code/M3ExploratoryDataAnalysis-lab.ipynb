## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
# Created a box plot for the 'ConvertedComp' column
plt.figure(figsize=(10, 6))
plt.boxplot(df['ConvertedComp'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Converted Compensation')
plt.xlabel('Converted Compensation (Annual Salary in USD)')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
# Calculatinh the first quartile and third quartile
Q1 = df['ConvertedComp'].quantile(0.25)  # 25th percentile
Q3 = df['ConvertedComp'].quantile(0.75)  # 75th percentile


IQR = Q3 - Q1

print(IQR)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound)
print(upper_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here

outliers = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)]

# Count the number of outliers
num_outliers = outliers.shape[0]

print(num_outliers)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# Filter the df to remove the outliers
new_df = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]

new_df.shape
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# Select only numeric columns from the df
numeric_df = df.select_dtypes(include='number')

# Calculating the correlation matrix for the numeric columns
correlation_matrix = numeric_df.corr()

# Extracting the correlation of Age column with other numerical columns
age_correlation = correlation_matrix['Age']

# Print the result
print("Correlation between Age and other numerical columns:")
print(age_correlation)
```
