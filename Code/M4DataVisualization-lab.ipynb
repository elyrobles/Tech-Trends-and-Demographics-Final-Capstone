Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Demo: How to run an sql query


```python
# Query to find the most common job type (DevType)
query = """
SELECT DevType, COUNT(*) as Count
FROM DevType
GROUP BY DevType
ORDER BY Count DESC
LIMIT 1
"""

# Execute the query and get the result
df_devtype_majority = pd.read_sql(query, conn)
print(df_devtype_majority)
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# Query to fetch the ConvertedComp column from the 'master' table
query = "SELECT ConvertedComp FROM master"

# Execute the query and load the df into a Pandas df
df = pd.read_sql(query, conn)

# Drop any rows with missing values in ConvertedComp
df = df.dropna(subset=['ConvertedComp'])

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=50, color='blue', edgecolor='black')
plt.title('Distribution of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
query = "SELECT Age FROM master" #get the Age column from the master table

# Execute the query and load the data into a Pandas df
df_age = pd.read_sql(query, conn)

# Dropping rows with missing values from Age
df_age = df_age.dropna(subset=['Age'])

# Plot the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(df_age['Age'])
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()


```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# get Age and WorkWeekHrs from the master table
query = "SELECT Age, WorkWeekHrs FROM master"

# Execute the query and load the data into a Pandas df
df_scatter = pd.read_sql(query, conn)

df_scatter = df_scatter.dropna(subset=['Age', 'WorkWeekHrs'])

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_scatter['Age'], df_scatter['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# get WorkWeekHrs, CodeRevHrs, and Age from the master table
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"

# Execute the query and load the data into a Pandas df
df_bubble = pd.read_sql(query, conn)

df_bubble = df_bubble.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Plot the bubble plot
plt.figure(figsize=(10, 6))
bubble_size = df_bubble['Age'] / 1.5  # Adjusts bubble size
plt.scatter(df_bubble['WorkWeekHrs'], df_bubble['CodeRevHrs'], s=bubble_size, alpha=0.5, color='blue', edgecolor='black')

# Add titles and labels
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Bubble size: Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)

# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# count the occurrences of each db in the 'DatabaseDesireNextYear' table
query = """
SELECT DatabaseDesireNextYear, COUNT(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5;
"""

# Execute the query and load the data into a Pandas df
df_pie = pd.read_sql(query, conn)


# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_pie['Count'], labels=df_pie['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# get WorkWeekHrs, CodeRevHrs, and Age from the master table
query = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Execute the query and load the data into a Pandas df
df_stacked = pd.read_sql(query, conn)


# Calculates the median of WorkWeekHrs and CodeRevHrs
median_workweekhrs = df_stacked['WorkWeekHrs'].median()
median_coderevhrs = df_stacked['CodeRevHrs'].median()

# Creates a df for plotting
data = pd.DataFrame({
    'Median WorkWeekHrs': [median_workweekhrs],
    'Median CodeRevHrs': [median_coderevhrs]
})

# Plot the stacked chart
data.plot(kind='bar', stacked=True, figsize=(8, 6), color=['blue', 'orange'])
plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
plt.xlabel('Age Group 30-35')
plt.ylabel('Hours')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# gets Age and ConvertedComp for respondents between the age of 45 and 60
query = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Execute the query and load the data into a Pandas df
df_line = pd.read_sql(query, conn)

df_line = df_line.dropna(subset=['ConvertedComp'])

# Group the data by Age and calculate the median ConvertedComp for each age
median_comp_by_age = df_line.groupby('Age')['ConvertedComp'].median()

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='blue')
plt.title('Median Converted Compensation by Age (45 to 60)')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# geth the MainBranch column from the master table
query = "SELECT MainBranch FROM master"

# Execute the query and load the data into a Pandas DataFrame
df_bar = pd.read_sql(query, conn)

# Count the occurrences of each unique value in MainBranch
mainbranch_counts = df_bar['MainBranch'].value_counts()

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of MainBranch Responses')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.grid(True)
plt.show()

```
